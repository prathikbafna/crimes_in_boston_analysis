```python
import pandas as pd
df=pd.read_csv("properdata.csv")
```

```python
df.head(6)
```

```python
df['REPORTING_AREA']=pd.to_numeric(df['REPORTING_AREA'],errors='corence') 
```

```python
df=df.fillna(method="ffill")
```

```python
df.to_csv("no_na_values.csv",index=False)
```

```python
df=df[df.YEAR==2018]
```

```python
df
```

```python
Lareny=0
Vandalism=0
Towed=0
AutoTheft=0
Robbery=0
DrugViolation=0
WarrantArrest=0

for i in df['OFFENSE_CODE_GROUP']:
    if(i=='Larceny'):
        Lareny=Lareny+1
    if(i=='Vandalism'):
        Vandalism=Vandalism+1
    if(i=='Towed'):
        Towed=Towed+1
    if(i=='Auto Theft'):
        AutoTheft=AutoTheft+1
    if(i=='Robbery'):
        Robbery=Robbery+1
    if(i=='Drug Violation'):
        DrugViolation=DrugViolation+1
    if(i=='Warrant Arrests'):
        WarrantArrest=WarrantArrest+1
```

```python
crime_numbers=[Lareny,Vandalism,Towed,AutoTheft,Robbery,DrugViolation,WarrantArrest]
crime_names=['Larency','Vandalism','Towed','AutoTheft','Robbery','DrugViolation','WarrantArrest']
```

```python
crime_numbers
```

```python
crime_names
```

```python
import matplotlib.pyplot as pls
```

```python
import numpy as np
```

```python
pb = pd.DataFrame(list(zip(crime_names,crime_numbers)), 
               columns =['crime', 'value']) 
```

```python
index=np.arange(len(crime_numbers))
pls.bar(index,crime_numbers)
pls.xlabel('crimes', fontsize=15)
pls.ylabel('No of crimes', fontsize=15) 
pls.xticks(index, crime_names, fontsize=10, rotation=30)
pls.show()
```

```python
pls.pie(crime_numbers,labels=crime_names,autopct='% 1.1f %%',shadow=True)
pls.title('CRIMES IN YEAR 2018')
pls.show()
```

```python
pls.barh(index,crime_numbers)
pls.ylabel("TYPES OF CRIMES",fontsize=15)
pls.xlabel("NUMBER OF OCCURANCE IN THE YEAR 2018",fontsize=10)
pls.yticks(index,crime_names,fontsize=10,rotation=30)
pls.title('CRIMES DETAILS')
pls.show()
```
